```python
# Dependencies
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
from nltk import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.corpus import stopwords
import pandas as pd
import numpy as np
import re    
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
Tweet = "CleanTrumpTweets.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
tweet = pd.read_csv(Tweet)
```

```python
tweet
```

```python
tweet_df = tweet[["absolute_tidy_tweets"]]
tweet_df.head()
```

```python
tweet_df['absolute_tidy_tweets'] = tweet_df['absolute_tidy_tweets'].astype(str)
```

```python
tweet_df['absolute_tidy_tweets'] = tweet_df['absolute_tidy_tweets'].str.lower()
```

```python
tweet_df
```

```python
tokens = tweet_df['absolute_tidy_tweets'].apply(word_tokenize)
```

```python
print(tokens)
```

```python
stop_words = set(stopwords.words("english"))
```

```python
print(stop_words)
```

```python
tweet_df['absolute_tidy_tweets'] = tweet_df['absolute_tidy_tweets'].apply(lambda x: " ".join(x for x in x.split() if x not in stop_words))
tweet_df['absolute_tidy_tweets'].head()
```

```python
freq = pd.Series(' '.join(tweet_df['absolute_tidy_tweets']).split()).value_counts()[:100]
freq
```

```python
st = PorterStemmer()
tweet_df['absolute_tidy_tweets'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

```python
from textblob import Word
tweet_df['absolute_tidy_tweets'] = tweet_df['absolute_tidy_tweets'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
tweet_df['absolute_tidy_tweets'].head()
```

```python
tweet_df['totalwords'] = tweet_df['absolute_tidy_tweets'].str.split().str.len()
```

```python
tweet_df
```

```python
total_words = tweet_df["totalwords"].sum()
```

```python
total_words
```

```python
freq = pd.Series(' '.join(tweet_df['absolute_tidy_tweets']).split()).value_counts()[:100]
freq 
```

```python
freq.to_frame()
```

```python
freq = pd.DataFrame(freq).reset_index()
freq.columns = ['Word', 'Frequency']
freq
```

```python
a=['amp' , 'u', 'get', 'would', 'also', 'call', 'ever', 'far', 'come', 'say', 'want', 'many', 'much', 'nan', 'well', 'even',
   'medium', 'like', 'nothing', 'done', 'dont', 'make', 'go', 'see', 'thing', 'way', 'back', 'must', ]

freq = freq[~freq['Word'].isin(a)]
```

```python
freq
```

```python
freq["tf"] = (freq["Frequency"] / total_words)
```

```python
reduced = freq["Frequency"].sum()
reduced
```

```python
freq["tf-red"] = (freq["tf"] / reduced)
freq
```

```python
count_row = freq.shape[0]  # gives number of row count
```

```python
freq["tf_df_row"] = (freq["Frequency"] / count_row)
freq                    
```

```python
freq.to_csv("Resources/tf_idf.csv", index=False, header=True)
```
