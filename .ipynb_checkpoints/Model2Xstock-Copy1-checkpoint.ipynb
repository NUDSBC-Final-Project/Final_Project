```python
import pandas as pd
import datetime
import joblib
```

```python
df = pd.read_csv('CleanTrumpTweets.csv')
df.head()
```

```python
df['date']=pd.to_datetime(df['tweetdate'])
df.head()
```

```python
tweetlist = list(df['tweetdate'].unique())
```

```python
len(tweetlist)
```

```python
data = {}
tempdate = []
temptweet = []

for tweet in tweetlist:

    df_ = df.loc[df['tweetdate'] == tweet][["tweetdate", "absolute_tidy_tweets"]]
    
    
    tweetsofday = ''
    for tidytweet in df_["absolute_tidy_tweets"]:
        tweetsofday = tweetsofday + str(tidytweet) + ' '
    
    tempdate.append(tweet)
    temptweet.append(tweetsofday)
    
```

```python
tempdate
```

```python
temptweet
```

```python
data ={
    "Date" : tempdate,
    "tweets" : temptweet
}
```

```python
df_tweets = pd.DataFrame(data)
df_tweets["Date"] = pd.to_datetime(df_tweets["Date"])
df_tweets.head()
```

```python
# Import csv
steel = pd.read_csv('Resources/X.csv')
steel['Date'] = pd.to_datetime(steel['Date'])
steel.head()
```

```python
merged_df = pd.merge(steel, df_tweets, how='inner', on='Date')
```

```python
df2=merged_df[['Date', 'tweets', 'P/N']]
df2.head()
```

```python
df2.shape
```

```python
from sklearn.model_selection import train_test_split
# splitting data for training and testing
X_train, X_test, y_train, y_test = train_test_split(df2['tweets'], df2['P/N'], test_size=.7 )
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(X_train)
```
