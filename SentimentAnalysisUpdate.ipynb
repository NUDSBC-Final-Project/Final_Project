```python
# Import tweet csv
tweets_df = pd.read_csv("CleanTrumpTweets.csv")
```

```python
tweets_df
```

```python
tweets_df["TWEETS"] = tweets_df['text'].astype(str)
```

```python
tweets = list(tweets_df["TWEETS"])
```

```python
#tweet_list=tweets.values.tolist()
```

```python
tweets[23]
```

```python
len(tweets)
```

```python
sentiment =[]
analyzer = SentimentIntensityAnalyzer()
for sentence in tweets:
    vs = analyzer.polarity_scores(sentence)
    sentiment.append(vs)
    print("{:-<65} {}".format(sentence, str(vs)))
```

```python
sentiment[5]
```

```python
sentiment_df = pd.DataFrame(sentiment)
```

```python
sentiment_df.shape
```

```python
df = pd.concat([tweets_df,sentiment_df],axis=1)
df.head()
```

```python
df_sent = df.groupby('tweetdate').agg({'compound': 'mean', 'neg': 'mean', 'neu': 'mean', 'pos': 'mean'}).reset_index()
df_sent.head()
```

```python
df_sent.to_csv('TweetSentiment.csv', index = False)
```

```python
df_total_sent = df.groupby('tweetdate').agg({'compound': 'sum', 'neg': 'sum', 'neu': 'sum', 'pos': 'sum'}).reset_index()
df_total_sent.head()
```

```python
df_total_sent.to_csv('TotalTweetSentiment.csv', index = False)
```
