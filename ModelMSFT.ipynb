```python
# Import csv
sentiment_df = pd.read_csv('TweetSentiment.csv')
sentiment_df.head()
```

```python
sentiment_df['tweetdate'] = pd.to_datetime(sentiment_df['tweetdate'])
```

```python
#df=merged_df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Future_Change', 'compound', 'neg', 'neu', 'pos', 'P/N']]
df=merged_df[['neg', 'neu', 'pos', 'P/N']]
df.head()
```

```python
# This is just a test to see correlation
# corr_df=merged_df[['Daily_Change', 'Future_Change', 'neg', 'neu', 'pos', 'compound']]
# corr_df.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
# One Hot encoding
# Step 0: Reformat data
data = df.values
X = data[:, 0:3]
y = data[:, 3]

# # Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
from keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_y = to_categorical(encoded_y)
```

```python
from sklearn.model_selection import train_test_split
# splitting data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, one_hot_y)
```

```python
from sklearn.preprocessing import StandardScaler
# Scaling data for use in the model
X_scaler = StandardScaler().fit(X_train)

# Scaling both training and testing
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
from tensorflow.keras.models import Sequential

deep_model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 3
number_hidden_nodes = 10
deep_model.add(Dense(units=number_hidden_nodes, activation='relu', input_dim=number_inputs))
deep_model.add(Dense(units=number_hidden_nodes, activation='relu'))
```

```python
number_classes = 2
deep_model.add(Dense(units=number_classes, activation='softmax'))
```

```python
deep_model.summary()
```

```python
deep_model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit (train) the model
history = deep_model.fit(
    X_train_scaled,
    y_train,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
import plotly.express as px
test_accuracy = model_accuracy
epochs = range(0,1000)
loss_train = history.history['acc']
fig = px.line(x=epochs, y=loss_train, title='Training Accuracy, Microsoft Corporation', range_y= [0.3,1])
fig.add_shape(
        # Line Horizontal
        go.layout.Shape(
            type="line",
            x0=0,
            y0=test_accuracy,
            x1=1000,
            y1=test_accuracy,
            line=dict(
                color="LightSeaGreen",
                width=4,
                dash="dashdot"
            )))
fig.update_layout(
    autosize=False,
    width=400,
    height=400,
    yaxis=go.layout.YAxis(
        title_text="Accuracy",
        tickmode="array",
        titlefont=dict(size=15),
    )
)

fig.update_layout(
    autosize=False,
    width=400,
    height=400,
    xaxis=go.layout.XAxis(
        title_text="Epoch",
        tickmode="array",
        titlefont=dict(size=15),
    )
)
fig.show()
```
